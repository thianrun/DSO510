```python
import pandas as pd
import statsmodels.formula.api as sm
import numpy as np
```

```python
data = pd.read_csv('hw4.csv')

data.head()
```

### Question 1

```python
data['ln_price'] = np.log(data['Price'])
data['ln_sales'] = np.log(data['Sales'])

data.head()
```

```python
result1 = sm.ols('ln_sales ~ ln_price', data = data).fit()

print(result1.summary())
```

### Question 2

```python
data['brand_1'] = [1 if i == 1 else 0 for i in data['Car_Brand']]
data['brand_2'] = [1 if i == 2 else 0 for i in data['Car_Brand']]
data['brand_3'] = [1 if i == 3 else 0 for i in data['Car_Brand']]

data.head()
```

```python
result2 = sm.ols('ln_sales ~ ln_price + Engine_size + Fuel_capacity + Fuel_efficiency + brand_1 + brand_2 + brand_3', data = data).fit()

print(result2.summary())
```

### Question 3

```python
result3 = sm.ols('Sales ~ Price + Engine_size + Fuel_capacity + Fuel_efficiency + brand_1 + brand_2 + brand_3', data = data).fit()

print(result3.summary())
```

### Question 4

```python
result4 = sm.ols('ln_sales ~ ln_price + Engine_size + Fuel_capacity + Fuel_efficiency + ln_price*brand_1 + ln_price*brand_2 + ln_price*brand_3', data = data).fit()

print(result4.summary())
```
