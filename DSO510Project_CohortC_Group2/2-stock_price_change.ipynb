```python
import pandas as pd
```

```python
stock_data = pd.read_excel('combined_data_source_categories.xlsx',index_col=0)
stock_data = stock_data.rename(columns={'Close':'Close_today'})
#stock_data['Close_tomorrow'] = [0 for i in stock_data['Close_today']]
stock_data.head()
```

```python
stock_data_for_cal = stock_data[["symbols","Date","Close_today"]].drop_duplicates().dropna().sort_values(["symbols","Date"],ascending=True)

stock_data_for_cal.head()
```

```python
list1 = []
temp=0
l = stock_data_for_cal['symbols'].unique()
for k in l:
    for i,j in enumerate(stock_data_for_cal['symbols']):
        if k==j and i>temp:
            temp=i
    list1.append(temp)    
            
```

```python
list1
```

```python
l=[]
current_index = 0
for i in stock_data_for_cal['symbols'].unique():
    for j,k in enumerate(stock_data_for_cal['symbols']):
        if k==i:
            if j < list1[current_index]:
                l.append(stock_data_for_cal.iloc[j+1]['Close_today'])
            elif j == list1[current_index]:
                l.append(stock_data_for_cal.iloc[j]['Close_today'])
    current_index = current_index + 1
            
```

```python
l
```

```python
stock_data_for_cal['Close_tomorrow'] = l
```

```python
stock_data_for_cal['Percent_price_change'] = 100*((stock_data_for_cal['Close_tomorrow'] - stock_data_for_cal['Close_today'])/stock_data_for_cal['Close_today'])
```

```python
stock_data_for_cal.head()
```

```python
stock_data = stock_data.drop(columns=['id','dayofweek'])
stock_data.head()
```

```python
stock_data = stock_data.merge(stock_data_for_cal[['Date','symbols','Close_tomorrow','Percent_price_change']], how = 'left', on = ['Date','symbols'])

stock_data.head()
```

```python
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
```

```python
stock_data['Day_of_week'] = stock_data['Date'].dt.dayofweek
```

```python
stock_data['Week_number'] = stock_data['Date'].dt.week
```

```python
stock_data['Month_number'] = stock_data['Date'].dt.month
```
